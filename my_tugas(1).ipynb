## Final Result Query

```python
FinalResult = %sql \
select customer.Id as CustomerId, \
coalesce(concat(Firstname, ' ',LastName),Firstname) as Name, \
Country, \
case when Phone is not null then 1 else 0 end as flag_phone, \
ifnull(Age,25) as Age, \
ifnull(max(datediff('2021-07-23',order_transaction.OrderDate)),0) as LOS, \
ifnull(sum(quantity),0) as TotalProduct, \
ifnull(sum(distinct order_transaction.TotalAmount),0) as TotalAmount, \
case when count(distinct PaymentType) = 1 and PaymentType = 'Cash' or PaymentType is null then 1 else 0 end as PaymentType \
from order_transaction \
left join orderitem on orderitem.OrderId = order_transaction.Id \
right join customer on customer.Id = order_transaction.CustomerId \
group by customer.Id \
order by PaymentType asc
```

```python
FinalResult = FinalResult.DataFrame()
```

```python
FinalResult.to_sql('finalresult',db_connection,if_exists='replace')
```
